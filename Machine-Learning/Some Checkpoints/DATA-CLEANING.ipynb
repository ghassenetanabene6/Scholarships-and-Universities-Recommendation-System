### Extraction of the University from Description

```python
# Function of extracting the university from the description text

def extract_univ(desc):
    l=[]
    for i in desc:
        l.append([X.text for X in nlp(str(i)).ents if (str(X.label_)=="ORG")])
    org=[]
    for i in l:
        if len(i)==0:
            org.append(None)
        elif (len(i)==1):
            if (len(i[0].split())<2):
                org.append(None)
            else:
                org.append(None)
        else:
            a=[]
            for x in i:
                if (len(x.split())>1) and (x not in a) and (("academy" in x.lower()) or ("college" in x.lower()) or ("university" in x.lower()) or ("school" in x.lower()) or ("institute" in x.lower())):
                    a.append(x)
            if len(a)==0:
                #a.append(max(list(set(i)),key=len))
                a.append(None)
            org.append(a[0])
            #org.append(max(a,key=len))
    return(org)
```

### Extraction of the Country from Description

```python
# Function of extracting the Country from the description text

def extract_location(desc):
    l=[]
    for i in desc:
        b=False
        a=[X.text for X in nlp(str(i)).ents if (str(X.label_)=="GPE")]
        if (a==[]):
            l.append(None)
        else: 
            for j in a:
                if (("usa" in j.lower()) or ("america" in j.lower())or (("united states")in j.lower()) or (("U.S.")in j) or (("US")in j)):
                    l.append("USA")
                    b=True
                    break
            if not b:
                if (len(set(a))==1):
                    l.append(a[0])
                else: l.append(", ".join(list(set(a))))
    return(l)
```

### Cleaning the Date Column

```python
# Function of cleaning the date

today = date.today()
def convert_date(date):
    L=list()
    y=([[X.text  for X in nlp(str(i)).ents if (str(X.label_)=="DATE")] for i in date])
    for d in y:
        if (d==[]):
            L.append(None)
            continue
        elif (len(str(d[-1]).split(" to "))==2): 
            w=str(d[-1]).split(" to ")[1].strip()
        elif (len(str(d[-1]).split(" and "))==2):
            w=str(d[-1]).split(" and ")[1].strip()
        elif (len(str(d[-1]).split("/"))==2):
            w=str(d[-1]).split("/")[1].strip()
        elif (len(str(d[-1]).split(" or"))==2):
            w=str(d[-1]).split(" or")[0].strip()
        elif (len(str(d[-1]).split(" of"))==2):
            w=str(d[-1]).split(" of")[1].strip()
        else: w=d[-1].replace("th",'').replace("spring",'').replace("(",'').replace(")",'').replace("1st",'1').strip()
        try:
            datetime_object = datetime.strptime(w, '%d %B %Y').date()
            L.append(datetime_object)
            continue
        except:
            pass
        try:
            datetime_object = datetime.strptime(w, '%B %d %Y').date()
            L.append(datetime_object)
            continue
        except:
            pass
        try:
            datetime_object = datetime.strptime(w, '%d %B').replace(year=today.year).date()
            L.append(datetime_object)
            continue
        except:
            pass
        try:
            datetime_object = datetime.strptime(w, '%B %d').replace(year=today.year).date()
            L.append(datetime_object)
            continue
        except:
            pass
        try:
            datetime_object = datetime.strptime(w, '%B %Y').date()
            L.append(datetime_object)
            continue
        except:
            pass
        try:
            datetime_object = datetime.strptime(w, '%B').replace(year=today.year).date()
            L.append(datetime_object)
            continue
        except:
            L.append(", ".join(d))
    return(L)
```

### Cleaning the text Column of df_model

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
import string

def text_process(text):
    stemmer = WordNetLemmatizer()
    nopunc = [char for char in text if char not in string.punctuation] #supprimer ponctuation
    nopunc = ''.join([i for i in nopunc if not i.isdigit()]) #supprimer les chiffres
    nopunc =  [word.lower() for word in nopunc.split() if word not in stopwords.words('english')] #supprimer les mots d'arrêts
    r = [stemmer.lemmatize(word) for word in nopunc]  #Racination des motes
    return(' '.join(r))
```

```python
for i,j in enumerate(df_model['Text']):
    df_model['Text'][i]=text_process(j)
```
